```python
prata_assunto = prata_assunto[['codigoAssuntoObra', 'assunto', 'sistema']][:]
prata_assunto = prata_assunto.query("sistema == 'Sisacoe'")
prata_assunto.info()
```

```python
sql_assunto = df_proc_sql_sisacoe.merge(prata_assunto, on='codigoAssuntoObra', how='left')
df_proc_sql = sql_assunto[['processo', 'sql_incra', 'codigoAssuntoObra', 'assunto', 'sistema_x']][:]
df_proc_sql.info()
```

```python
df_proc_sql['sql_incra'] = df_proc_sql['sql_incra'].astype(str)
```

```python
df_proc_sql['sql_incra'] = df_proc_sql['sql_incra'].apply(lambda x: x.replace('.', '').replace('-', ''))
df_proc_sql.head()
```

```python
df_proc_sql['sql_incra'] = df_proc_sql['sql_incra'].apply(lambda x: x.strip())
df_proc_sql.head()
```

### Arquivo Busca

```python
with open('sqls_metro.txt') as sqls_file:
    consulta = sqls_file.readlines()

consulta = list(map(lambda x: x[:-1], consulta))
len(consulta)
```

```python
resultado = [df_proc_sql.query(f"sql_incra == '{sql}'") for sql in consulta]
result_df = pd.concat(resultado)
result_df.info()
```

```python
result_df.reset_index(drop=True, inplace=True)
```

```python
result_df.tail(60)
```

```python
result_df = result_df.drop_duplicates()
result_df.info()
```

```python
result_df['processo'] = result_df['processo'].apply(lambda x: x.replace('.', '').replace('/', '').replace('-', ''))
result_df.head()
```
