## Busca SQL MP

Método:

Para cara arquivo contendo dados dos diversos sistemas, gerar dataframe com número de processo e SQL relacionado (1 para n), unir os data frames em um único e após, iterar sobre a lista de sqls que se quer pesquisar e procurar, esse determinado sql, trazendo todos os processos que forem encontrados.

```python
import pandas as pd
```

### Aprova Digital

```python
ad = pd.read_csv('ad_localizacoes.csv', encoding='latin1', sep=';')
ad.info()
```

```python
ad = ad[['n_processo', 'sql']]
ad.head(20)
```

```python
ad['sql'] = ad['sql'].astype(str)
```

```python
# Retirando as mascaras
def no_mask(text):
    return text.replace('.', '').replace('-', '').zfill(11)

ad['sql_clean'] = ad['sql'].apply(no_mask)
ad.head(20)
```

```python
aprova_digital = ad[['n_processo', 'sql_clean']]
aprova_digital
```

### Portal

```python
portal = pd.read_csv('portal_sqls.csv', sep=',')
portal.info()
```

```python
portal['sql_incra'] = portal['sql_incra'].astype(str)
```

```python
portal['sql_incra']
```

```python
def portal_clean(valor):
    return valor[:-2].zfill(11)
```

```python
portal['sql_incra'] = portal['sql_incra'].apply(portal_clean)
portal['sql_incra']
```

```python
portal.columns = ['n_processo', 'sql_clean']
```

### Sisacoe

```python
sisacoe = pd.read_csv('sisacoe_sqls.csv', sep=',')
sisacoe.info()
```

```python
sisacoe['sql_incra'] = sisacoe['sql_incra'].apply(lambda x: str(x).zfill(11))
sisacoe.head()
```

```python
sisacoe.columns = ['n_processo', 'sql_clean']
sisacoe.head()
```

### SLC2

```python
slc2_2017 = pd.read_excel('slcii_2017_2019.xlsx')
slc2_2017.info()
```

```python
slc2_2020 = pd.read_excel('slcii_2020_2023.xlsx')
slc2_2020.info()
```

```python
slc2 = pd.concat([slc2_2017, slc2_2020])
slc2.info()
```

```python
slc2.head()
```

```python
slc2 = slc2[['Número Processo', 'Número SQL']]
slc2.head()
```

```python
slc2.columns = ['n_processo', 'sql_clean']
slc2['sql_clean'] = slc2['sql_clean'].apply(no_mask)
slc2.head()    
```
