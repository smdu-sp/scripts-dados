```python
import pandas as pd
```

## Busca por SQL

### Aprova Digital

```python
ad = pd.read_csv('ad_localizacoes.csv', encoding='latin1', sep=';')
ad.info()
```

```python
loc = ad[['sql', 'n_processo', 'assunto', 'logradouro', 'numero', 'bairro']][:]
loc.head(20)
```

```python
loc['sql'] = loc['sql'].astype(str)
```

```python
# Retirando as mascaras
def no_mask(text):
    return text.replace('.', '').replace('-', '').zfill(11)

loc['sql_clean'] = loc['sql'].apply(no_mask)
loc.head(20)
```

```python
localizacao = loc[['sql_clean', 'n_processo', 'assunto', 'logradouro', 'numero', 'bairro']][:]
localizacao
```

```python
localizacao['sql_list'] = localizacao['sql_clean'].apply(lambda x: x.replace(' ', '').split(','))
localizacao.head()
```

```python
localizacao = localizacao.explode('sql_list')
localizacao.shape
```

```python
localizacao_ex = localizacao[['sql_list', 'n_processo', 'assunto', 'logradouro', 'numero', 'bairro']]
localizacao_ex.head()
```

```python
localizacao_ex.columns = ['sql_list', 'processo', 'assunto', 'logradouro', 'numero', 'bairro']
localizacao_ex.info()
```

### AD - outras tabelas

```python
# Balanco processos - tabela pronta com diversos dos campos necessários à busca
balanco_raw = pd.read_csv('ad_balanco_processo.csv', encoding='latin1', sep=';')
balanco_raw.info()
```

```python
balanco = balanco_raw[['processo', 'sistema_origem', 'data_autuacao', 'situacao', 'proprietario']]
balanco.head()
```

```python
# tabelas despachos
despachos_raw = pd.read_csv('ad_dispatched_despachos.csv', encoding='latin1', sep=';')
despachos_raw.info()
```

```python
despachos = despachos_raw[['processo', 'data_publicacao', 'documento']][:]
despachos.head()
```

```python
despachos['data_emissao_alvara'] = despachos['data_publicacao']
despachos.head()
```

### Merge

```python
ad_temp = pd.merge(localizacao_ex, balanco, how='left', on='processo')
ad_merge = pd.merge(ad_temp, despachos, how='left', on='processo')
ad_merge.info()
```

```python
ad_merge['cod_assunto_obra'] = None
ad_merge['cod_assunto'] = None
ad_merge['cod_aditivo'] = None
ad_merge.info()
```

```python
ad_final = ad_merge[['sistema_origem', 'processo', 'data_autuacao', 'cod_assunto_obra', 'sql_list', 'cod_assunto', 'cod_aditivo', 'assunto', 'situacao',
                     'data_publicacao' ,'documento', 'data_emissao_alvara', 'proprietario', 'logradouro', 'numero', 'bairro']]
ad_final.rename(columns={'sql_list': 'sql', 'situacao': 'status', 'proprietario': 'interessados'}, inplace=True)
ad_final.info()
```

### Arquivo Busca

```python
with open('sqls_metro.txt') as sqls_file:
    consulta = sqls_file.readlines()

consulta = list(map(lambda x: x[:-1], consulta))
len(consulta)
```

```python
resultado = [ad_final.query(f"sql == '{sql}'") for sql in consulta]
result_df = pd.concat(resultado)
result_df.head(20)
```

```python
result_df.reset_index(drop=True, inplace=True)
```

```python
result_df.info()
```

```python
result_df = result_df.drop_duplicates()
result_df
```

```python
result_df.to_excel('_consulta_ad_sql_mp_metro.xlsx', index=False)
```

## Busca por endereço

### Preparação do arquivo de endereços com chaves de busca

```python
# dados endereco
with open('enderecos_metro.txt', 'r', encoding='utf-8') as end_txt:
    nao_formatado = end_txt.readlines()
nao_formatado = list(map(lambda x: x[:-1], nao_formatado))
nao_formatado[:3]
```

```python
# tranformando em dataframe
df_end = pd.DataFrame({'endereco_nao_formatado': nao_formatado})
df_end
```

```python
df_end['endereco_nf_min'] = df_end['endereco_nao_formatado'].str.lower()
df_end.head(3)
```

```python
# Tirando alguns titulos que podem ser inconsistentes
to_replace = (('engenheiro ', ''), ('doutor ', ''), ('dr. ', ''), ('professor', ''), ('santa ', ''), ('conde ', ''), ('coronel', ''))
for k, i in to_replace:
    df_end['endereco_nf_min'] = df_end['endereco_nf_min'].str.replace(k, i)
df_end.tail()
```

```python
df_end['chave_pesquisa']  = df_end['endereco_nf_min'].apply(lambda x: x[4:12])
#list(df_end['chave_pesquisa'])
```

```python
df_end['numero_nf'] = df_end['endereco_nf_min'].apply(lambda x: x[x.find(', n') + 3: x.find('-')])
df_end['numero_nf']
```

```python
df_end['numero'] = df_end['numero_nf'].str.strip()
df_end['numero']
```

```python
df_end_pesquisa = df_end[['chave_pesquisa', 'numero']]
list(df_end_pesquisa['chave_pesquisa'])
df_end_pesquisa
```

```python
for index, row in df_end_pesquisa.iterrows():
    print(row['chave_pesquisa'], row['numero'])
```

### Busca na tabela

```python
ad_final['sq'] = ad_final['sql'].apply(lambda x: x[:6])
ad_final['sq']
```

```python
consulta_sq = list(map(lambda x: x[:6], consulta))
consulta_sq
```

```python
ad_final['logradouro'] = ad_final['logradouro'].astype(str)
ad_final['numero'] = ad_final['numero'].astype(str)
```

```python
ad_final['logradouro_formatado'] = ad_final['logradouro'].apply(lambda x: str(x).lower())
```

```python
ad_filtro_sq = ad_final.query("sq.isin(@consulta_sq)")[:]
```

```python
result = []
for index, row in df_end_pesquisa.iterrows():
    temp = ad_filtro_sq.query(f"(logradouro_formatado.str.contains('{row['chave_pesquisa']}')) and (numero.str.contains('{row['numero']}'))")
    result.append(temp)
final_result = pd.concat(result)
final_result.shape
```

```python
final_result.to_csv('temp.csv', sep=';', index=False)
```
