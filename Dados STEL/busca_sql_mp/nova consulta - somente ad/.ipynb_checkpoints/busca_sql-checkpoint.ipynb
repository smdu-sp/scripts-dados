## Busca SQL MP

Método:

Para cara arquivo contendo dados dos diversos sistemas, gerar dataframe com número de processo e SQL relacionado (1 para n), unir os data frames em um único e após, iterar sobre a lista de sqls que se quer pesquisar e procurar, esse determinado sql, trazendo todos os processos que forem encontrados.

```python
import pandas as pd
```

### Aprova Digital

```python
ad = pd.read_csv('ad_localizacoes.csv', encoding='latin1', sep=';')
ad.info()
```

```python
ad = ad[['n_processo', 'sql']]
ad.head(20)
```

```python
ad['sql'] = ad['sql'].astype(str)
```

```python
# Retirando as mascaras
def no_mask(text):
    return text.replace('.', '').replace('-', '').zfill(11)

ad['sql_clean'] = ad['sql'].apply(no_mask)
ad.head(20)
```

```python
aprova = ad[['n_processo', 'sql_clean']]
aprova
```

```python
aprova = aprova.assign(sql_clean=aprova['sql_clean'].str.split(';')).explode('sql_clean')
```

```python
aprova.head(20)
```

```python
aprova['origem'] = 'Aprova Digital'
aprova.head()
```

### Portal

```python
portal = pd.read_csv('portal_sqls.csv', sep=',')
portal.info()
```

```python
portal['sql_incra'] = portal['sql_incra'].astype(str)
```

```python
portal['sql_incra']
```

```python
def portal_clean(valor):
    return valor[:-2].zfill(11)
```

```python
portal['sql_incra'] = portal['sql_incra'].apply(portal_clean)
portal['sql_incra']
```

```python
portal.columns = ['n_processo', 'sql_clean']
```

```python
portal['origem'] = 'portal'
portal.head()
```

### Sisacoe

```python
sisacoe = pd.read_csv('sisacoe_sqls.csv', sep=',')
sisacoe.info()
```

```python
sisacoe['sql_incra'] = sisacoe['sql_incra'].apply(lambda x: str(x).zfill(11))
sisacoe.head()
```

```python
sisacoe.columns = ['n_processo', 'sql_clean']
sisacoe.head()
```

```python
sisacoe['origem'] = 'sisacoe'
sisacoe.head()
```

### SLC2

```python
slc2_2017 = pd.read_excel('slcii_2017_2019.xlsx')
slc2_2017.info()
```

```python
slc2_2020 = pd.read_excel('slcii_2020_2023.xlsx')
slc2_2020.info()
```

```python
slc2 = pd.concat([slc2_2017, slc2_2020])
slc2.info()
```

```python
slc2.head()
```

```python
slc2 = slc2[['Número Processo', 'Número SQL']]
slc2.head()
```

```python
slc2.columns = ['n_processo', 'sql_clean']
slc2['sql_clean'] = slc2['sql_clean'].apply(no_mask)
slc2.head()    
```

```python
slc2['origem'] = 'slc2'
slc2.head()
```

### SLCe

```python
from bs4 import BeautifulSoup
import os
```

```python
def table_reader(filename):
    
    with open(filename, 'r') as strange_file:
        file = strange_file.read()
        
    strange_symbol = file[3]
    cleaned_file = file.replace(strange_symbol, '')

    soup = BeautifulSoup(cleaned_file, 'html.parser')
    
    table_container = [[i.text for i in row.find_all('td')] for row in soup.find_all('tr')]

    return pd.DataFrame(table_container[1:], columns=table_container[0])
```

```python
dataframes = [table_reader(i) for i in os.listdir() if i.endswith('.xls')]
```

```python
slce_full = pd.concat(dataframes)
slce_full.info()
```

```python
slce = slce_full[['Número do Processo (Número do Protocolo)', 'Número da Localização']]
slce.info()
```

```python
slce.columns = ['n_processo', 'sql_clean']
```

```python
slce.info()
```

```python
slce.head()
```

```python
slce['sql_clean'] = slce['sql_clean'][:].apply(no_mask)
slce.head(20)
```

```python
slce['origem'] = 'slce'
slce.head()
```

### Tabelas Unificadas

```python
sistemas_list = [aprova, portal, sisacoe, slc2, slce]
sistemas = pd.concat(sistemas_list)
sistemas.info()
```

```python
sistemas.head()
```

```python
sistemas.query("sql_clean == '04104300251'")
```

### Arquivo Busca

```python
with open('sqls_metro.txt') as sqls_file:
    consulta = sqls_file.readlines()

consulta = list(map(lambda x: x[:-1], consulta))
len(consulta)
```

```python
resultado = [sistemas.query(f"sql_clean == '{sql}'") for sql in consulta]
result_df = pd.concat(resultado)
result_df.head(20)
```

```python
result_df.reset_index(drop=True, inplace=True)
```

```python
result_df.head()
```

```python
result_df = result_df.drop_duplicates()
result_df
```

```python
result_df = result_df[['sql_clean', 'n_processo', 'origem']]
result_df.head()
```

```python
result_df.columns = ['sql', 'processo', 'origem']
result_df.tail()
```

```python
result_df.to_excel('_consulta_sql_mp_metro.xlsx', index=False)
```
