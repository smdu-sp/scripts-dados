## Aprova Digital - Planurb

```python
# PLanurb
df_planurb = pd.read_excel('dados_entrada\planurbV2_2024-04-24T12-24-44_prod.xlsx')
df_planurb.info()
```

```python
# Limpando a quantidade de colunas
df_planurb_colunas = df_planurb[['Assunto', 'Número do processo', 'Número SEI', 'Data de Deferimento', 'Status', 'N° de Unidades HIS' ,'N° de Unidades HIS1', 'N° de Unidades HIS2', 'N° de Unidades HMP', 'O Processo foi reaberto?', 'Data de Finalização']]
df_planurb_colunas.head()
```

```python
df_planurb_colunas['Status'].value_counts()
```

```python
# Arrumando nome das colunas

df_planurb_assuntos = df_planurb_colunas
df_planurb_assuntos.columns = ['assunto', 'protocolo', 'n_processo', 'data_deferimento', 'status', 'qtd_his', 'qtd_his1', 'qtd_his2', 'qtd_hmp', 'O Processo foi reaberto?', 'Data de Finalização']
df_planurb_assuntos.head()
```

```python
# carregando Assuntos a serem filtrados
with open('assuntos.txt', 'r', encoding= "utf-8") as txtfile:
    lista_assuntos = txtfile.readlines()
    lista_assuntos = list(map(lambda x: x[:-1], lista_assuntos))

lista_assuntos
```

```python
# Aplicando filtro de assuntos
df_planurb_assuntos_filtrados = df_planurb_assuntos
df_planurb_assuntos_filtrados = df_planurb_assuntos.query("assunto in @lista_assuntos")
df_planurb_assuntos_filtrados.info()
```

```python
# conversão colunas de valores para int

df_planurb_int = df_planurb_assuntos_filtrados[:]
df_planurb_int[['qtd_his', 'qtd_his1', 'qtd_his2', 'qtd_hmp']] = df_planurb_int[['qtd_his', 'qtd_his1', 'qtd_his2', 'qtd_hmp']].apply(pd.to_numeric, errors='coerce')
df_planurb_int[['qtd_his', 'qtd_his1', 'qtd_his2', 'qtd_hmp']] = df_planurb_int[['qtd_his', 'qtd_his1', 'qtd_his2', 'qtd_hmp']].fillna(0).astype(int)
df_planurb_int.info()

```

```python
# Separando 2 dataframes, um contendo processos deferidos nunca reabertos e outro contendo processos deferidos reabertos

# Filtrando apenas processos deferidos nunca reabertos e com data de finalização nula
df_planurb_def = df_planurb_int
df_planurb_def = df_planurb_int[(df_planurb_int['status'] == 'Deferido') & 
                                     (df_planurb_int['O Processo foi reaberto?'] == 'Não') & 
                                     (df_planurb_int['Data de Finalização'].isnull())]


# Filtrando apenas processos deferidos, reabertos e com data de finalização não nula
df_planurb_reab = df_planurb_int[(df_planurb_int['status'] == 'Deferido') & 
                                     (df_planurb_int['O Processo foi reaberto?'] == 'Sim') & 
                                     (~df_planurb_int['Data de Finalização'].isnull())]
df_planurb_reab.info()
```

```python
df_planurb_def.info()
```

```python
# criação coluna his_soma
df_planurb_his_total = df_planurb_def.copy()  
df_planurb_his_total.loc[:, 'his_total'] = df_planurb_his_total['qtd_his'] + df_planurb_his_total['qtd_his1'] + df_planurb_his_total['qtd_his2']
df_planurb_his_total = df_planurb_his_total.loc[:, ['assunto', 'protocolo', 'n_processo', 'data_deferimento', 'status', 'qtd_his', 'qtd_his1', 'qtd_his2', 'his_total', 'qtd_hmp', 'O Processo foi reaberto?', 'Data de Finalização']]

# criação coluna his_soma no dataframe de reabertos
df_planurb_his_total_reab = df_planurb_reab.copy()  
df_planurb_his_total_reab.loc[:, 'his_total'] = df_planurb_his_total_reab['qtd_his'] + df_planurb_his_total_reab['qtd_his1'] + df_planurb_his_total_reab['qtd_his2']
df_planurb_his_total_reab = df_planurb_his_total_reab.loc[:, ['assunto', 'protocolo', 'n_processo', 'data_deferimento', 'status', 'qtd_his', 'qtd_his1', 'qtd_his2', 'his_total', 'qtd_hmp', 'O Processo foi reaberto?', 'Data de Finalização']]
```

```python
df_planurb_his_total_reab.info()
```

```python
# convertendo para date time a coluna data_deferimento
df_planurb_data = df_planurb_his_total[:]
df_planurb_data['data_deferimento'] = pd.to_datetime(df_planurb_data['data_deferimento'], dayfirst=True)


# convertendo para date time a coluna data de finalização
df_planurb_data_reab = df_planurb_his_total_reab[:]
df_planurb_data_reab['Data de Finalização'] = pd.to_datetime(df_planurb_data_reab['Data de Finalização'], dayfirst=True)
```

```python
# criação do dataframe clean para futura conferencia
df_planurb_clean = df_planurb_data.copy()
df_planurb_clean_reab= df_planurb_data_reab.copy()
```

```python
# agrupando por mês, utilizando data_deferimento para os processos deferidos nunca reabertos
grouper = pd.Grouper(key='data_deferimento', freq='MS', label='left')
df_planurb_agrupado = df_planurb_data.groupby(grouper)
df_planurb_agrupado = df_planurb_agrupado[['his_total', 'qtd_hmp']].sum()

# agrupando por mês, utilizando Data de Finalização para os processos deferidos reabertos
grouper = pd.Grouper(key='Data de Finalização', freq='MS', label='left')
df_planurb_agrupado_reab = df_planurb_data_reab.groupby(grouper)
df_planurb_agrupado_reab = df_planurb_agrupado_reab[['his_total', 'qtd_hmp']].sum()
```

```python
# data frame final agrupado dos processos reabertos
df_planurb_agrupado_reab
```

```python
# data frame final agrupado dos processos nunca reabertos
df_planurb_agrupado
```

```python
# agrupando df_planurb_agrupado e df_planurb_agrupado_reab em um único dataframe
df_planurb_agrupado_final = pd.concat([df_planurb_agrupado, df_planurb_agrupado_reab], axis=1)

# convertendo nulls para 0 e convertendo para int
df_planurb_agrupado_final = df_planurb_agrupado_final.fillna(0).astype(int)

# renomeando coluna 3 para "his_total_reaberto" e coluna 4 para "qtd_hmp_reaberto"
df_planurb_agrupado_final.columns = ['his_total_nunca_reaberto', 'qtd_hmp_nunca_reaberto', 'his_total_reaberto', 'qtd_hmp_reaberto']

# criação de uma nova coluna chamado his_total_final e qtd_hmp_final que é a soma das colunas his_total e his_total_reaberto e qtd_hmp e qtd_hmp_reaberto
df_planurb_agrupado_final['his_total_final'] = df_planurb_agrupado_final['his_total_nunca_reaberto'] + df_planurb_agrupado_final['his_total_reaberto']
df_planurb_agrupado_final['qtd_hmp_final'] = df_planurb_agrupado_final['qtd_hmp_nunca_reaberto'] + df_planurb_agrupado_final['qtd_hmp_reaberto']

# reordenando as colunas
df_planurb_agrupado_final = df_planurb_agrupado_final[['his_total_nunca_reaberto', 'his_total_reaberto', 'his_total_final', 'qtd_hmp_nunca_reaberto', 'qtd_hmp_reaberto', 'qtd_hmp_final']]
```

```python
df_planurb_agrupado_final
```

```python
# dataframe final dados pré agrupamento
df_planurb_clean
```

```python
# dataframe final dados pré agrupamento reabertos
df_planurb_clean_reab
```

```python
df_planurb_agrupado_final.to_excel('dados_saida/ad_his_hmp_mes_v2.xlsx', index = True)
df_planurb_clean.to_excel('dados_saida/ad_lista_processos_deferidos.xlsx', index = False)
df_planurb_clean_reab.to_excel('dados_saida/ad_lista_processos_reabertos.xlsx', index = False)
```

## SISSEL

```python
# SISSEL
df_sissel = pd.read_excel('dados_entrada\AlvaraseAutos.xlsx')
df_sissel['his_total'] = None
df_sissel.head()
```

```python
# Limpando a quantidade de colunas
df_sissel_colunas = df_sissel[['Assunto', 'Processo', 'Data Emissão', 'his_total' ,'HIS1', 'HIS2', 'HMP']]
df_sissel_colunas.head()
```

```python
df_sissel_colunas['Assunto'].value_counts()
```

```python
df_sissel_colunas.info()
```

```python
# arrumando nome das colunas
df_sissel_colunas.columns = ['assunto', 'n_processo', 'data_deferimento', 'his_total', 'qtd_his1', 'qtd_his2', 'qtd_hmp']
df_sissel_colunas.head()
```

```python
# transformando em date time
df_sissel_data = df_sissel_colunas[:]
df_sissel_data['data_deferimento'] = pd.to_datetime(df_sissel_data['data_deferimento'], dayfirst=True)
df_sissel_data.info()
```

```python
# remocao dos nans e conversao dos valores float para int
df_sissel_conv = df_sissel_data
df_sissel_conv = df_sissel_conv.fillna(0)
df_sissel_conv['his_total'] = df_sissel_conv['his_total'].astype(int)
df_sissel_conv['qtd_his1'] = df_sissel_conv['qtd_his1'].astype(int)
df_sissel_conv['qtd_his2'] = df_sissel_conv['qtd_his2'].astype(int)
df_sissel_conv['qtd_hmp'] = df_sissel_conv['qtd_hmp'].astype(int)

soma_his1 = df_sissel_conv['his_total'].sum()
soma_his1 = df_sissel_conv['qtd_his1'].sum()
soma_his2 = df_sissel_conv['qtd_his2'].sum()
soma_hmp = df_sissel_conv['qtd_hmp'].sum()

soma_his1, soma_his2, soma_hmp
```

```python
# criação coluna his_soma

df_sissel_soma = df_sissel_conv.copy()  
df_sissel_soma.loc[:, 'his_total'] = df_sissel_soma['qtd_his1'] + df_sissel_soma['qtd_his2']
df_sissel_soma = df_sissel_soma.loc[:, ['assunto', 'n_processo', 'data_deferimento', 'qtd_his1', 'qtd_his2', 'his_total','qtd_hmp']]
```

```python
# criação do dataframe clean para futura conferencia

df_sissel_clean = df_sissel_soma.copy()
```

```python
# agrupando por mês
grouper = pd.Grouper(key='data_deferimento', freq='MS', label='left')
df_sissel_agrupado = df_sissel_clean.groupby(grouper)
df_sissel_agrupado = df_sissel_agrupado[['his_total', 'qtd_hmp']].sum()
```
