```python
import pandas as pd
from bs4 import BeautifulSoup
import os
```

```python
def table_reader(filename):
    
    with open(filename, 'r') as strange_file:
        file = strange_file.read()
        
    strange_symbol = file[3]
    cleaned_file = file.replace(strange_symbol, '')

    soup = BeautifulSoup(cleaned_file, 'html.parser')
    
    table_container = [[i.text for i in row.find_all('td')] for row in soup.find_all('tr')]

    return pd.DataFrame(table_container[1:], columns=table_container[0])
```

```python
#erro
#df1 = pd.read_excel('busca_sql_2012.xls', engine='xlrd')
#df1.info()
```

```python
# Criando um dataframe para cada arquivo
dataframes = [table_reader(i) for i in os.listdir() if i.endswith('.xls')]
```

```python
# Concatenando os dataframes a unindo os dados
df = pd.concat(dataframes)
df.info()
```

```python
df.columns = ['sql', 'n_processo', 'n_protocolo', 'assunto', 'situacao', 'ano', 'competencia', 'unidade_administrativa']
df.head(20)
```
