### Consulta por endereço

#### Busca por codlog

```python
df_codlog = df_endereco.dropna(subset=['numero', 'codlog', 'processo'])
df_codlog.loc[:,'processo'] = df_codlog['processo'].replace('', None)
```

```python
# Pradronização codlogs da base BI
def codlog_nomask(codlog):
    codlog = codlog.replace('-', '')  # Tirando o hifen do digito verificador
    codlog = codlog.zfill(6)  # Todos com 6 dígitos 

    return codlog
    
df_codlog.loc[:,'codlog_pdr'] = df_codlog['codlog'].apply(codlog_nomask)
df_codlog.head(10)
```

```python
# Padronização dos números
def not_a_num(string):
    for n, i in enumerate(string):
        if not i.isnumeric():
            return n
    return n
    
def num_cleaner(num):
    num = str(num) + 'a'
    return num[:not_a_num(num)]

def make_int(num):
    return int(num)

df_codlog.loc[:,'numero_pdr'] = df_codlog['numero'].apply(num_cleaner)
df_codlog.loc[:,'numero_pdr'] = df_codlog['numero_pdr'].replace('', 0)
df_codlog.loc[:,'numero_pdr'] = df_codlog['numero_pdr'].apply(make_int)
df_codlog.head()
```

```python
# Lendo codlogs e numeros
with open('filtro_endereco.txt', 'r') as txt_file:
    endereco_list = txt_file.readlines()

endereco_list = [i[:-1] for i in endereco_list]
codlog_num = [(i.split(',')[0], int(i.split(',')[2]), int(i.split(',')[3])) for i in endereco_list]

codlog_num
```

```python
df_container = [df_codlog[(df_codlog['codlog'] == i[0]) & (df_codlog['numero_pdr'] > i[1]) & (df_codlog['numero_pdr'] < i[2])] for i in codlog_num]

#consulta_codlog = df_codlog[(df_codlog['codlog'] == '009997') & (df_codlog['numero_pdr'] > 1000) & (df_codlog['numero_pdr'] < 1500)]
consulta_codlog = pd.concat(df_container)  
consulta_codlog
```

```python
# left join processos on processo
df_merge_codlog = pd.merge(consulta_codlog, df_assunto, how='left', on='processo')
df_merge_codlog = pd.merge(df_merge_codlog, df_sql_incra, how='left', on='processo')
df_merge_codlog.info()
```

```python
# Filtrando por assunto
assunto_filtrado_codlog = df_merge_codlog[df_merge_codlog['assuntoFormatado'].isin(assunto_list)]
assunto_filtrado_codlog['consulta'] = 'codlog'
```

```python
consulta_final_codlog = assunto_filtrado_codlog[['sql_incra', 'processo', 'assuntoFormatado', 'situacaoAssunto', 'dtInclusaoAssunto', 'dtEmissaoAlvara', 'numeroAlvara', 'codlog_pdr', 'logradouro', 'numero_pdr', 'sistema_x', 'consulta']]
consulta_final_codlog.columns = ['sql_incra', 'processo', 'assunto', 'situacao', 'data_inclusao_assunto', 'data_emissao_alvara', 'numero_alvara', 'codlog', 'logradouro', 'numero', 'sistema_origem', 'consulta']
consulta_final_codlog = consulta_final_codlog.sort_values(by=['codlog', 'data_inclusao_assunto'], ascending=[True, False])
consulta_final_codlog = consulta_final_codlog.drop_duplicates(subset=['processo', 'assunto'])
```
